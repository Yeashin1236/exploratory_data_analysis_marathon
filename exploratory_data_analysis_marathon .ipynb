```python
# Rename Cloumns
marathon2 = marathon2.rename(columns= { 'Year of event' : 'year',
                                       'Event dates' : 'race_day' ,
                                       'Event name' : 'race_name',	
                                       'Event distance/length' : 'race_length',	
                                       'Event number of finishers' : 'race_number_of_finisher',	
                                       'Athlete performance': 'athlete_performance',
                                       'Athlete gender':'athlete_gender',
                                       'Athlete average speed':'athlete_average_speed',	
                                       'Athlete ID':'athlete_iD',	
                                    
})
```

```python
# Difference in speed for the 50k, 50mi, male to female
marathon3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age group are the best in the 50m race (20 + race min)
marathon3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
```

```python
# What age group are the worst in the 50m race (10 + race min)
marathon3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9')
```

```python
# Seasons for the data -> slower in summer than winter?
# Spring 3-5
# Summer 6-8
# Fall 9-11
# Winter 12-2
# Split between two decimals
```

```python
marathon3['race_month'] =  marathon3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
marathon3['race_season'] = marathon3['race_month'].apply(
    lambda x: 'Winter' if x > 11 else 
              'Fall' if x > 8 else 
              'Summer' if x > 5 else 
              'Spring' if x > 2 else 
              'Winter'
)
```

```python
marathon3["race_season"]
```

```python
marathon3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```

```python
# 50 miler only
marathon3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```
